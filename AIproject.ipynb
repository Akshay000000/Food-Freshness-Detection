```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ulnnproject/food-freshness-dataset")

print("Path to dataset files:", path)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
from PIL import Image

# EXTREME SPEED Configuration
IMG_SIZE = 96
BATCH_SIZE = 128
EPOCHS = 10
LEARNING_RATE = 0.001
SUBSET_SIZE = 10000  # Use only 10k images for blazing fast training!

print("="*60)
print("BLAZING FAST Food Freshness Classifier")
print("Using dataset subset for ultra-fast training")
print("="*60)

# Dataset paths
dataset_path = "/root/.cache/kagglehub/datasets/ulnnproject/food-freshness-dataset/versions/1"
data_dir = "./Dataset_flat"

# Enable mixed precision
from tensorflow.keras import mixed_precision
mixed_precision.set_global_policy('mixed_float16')

# Load images into memory for faster training
print("\nLoading subset of data into memory...")

X_data = []
y_data = []
class_names = sorted([d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))])
class_to_idx = {name: idx for idx, name in enumerate(class_names)}

images_per_class = SUBSET_SIZE // len(class_names)

for class_name in class_names:
    class_path = os.path.join(data_dir, class_name)
    images = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]

    # Take only subset from each class
    selected = images[:images_per_class]

    for img_name in selected:
        img_path = os.path.join(class_path, img_name)
        try:
            img = Image.open(img_path).convert('RGB')
            img = img.resize((IMG_SIZE, IMG_SIZE))
            img_array = np.array(img) / 255.0
            X_data.append(img_array)
            y_data.append(class_to_idx[class_name])
        except:
            continue

X_data = np.array(X_data, dtype=np.float32)
y_data = np.array(y_data)

print(f"Loaded {len(X_data)} images")
print(f"Classes: {len(class_names)}")

# Split data
X_train, X_val, y_train, y_val = train_test_split(
    X_data, y_data, test_size=0.2, random_state=42, stratify=y_data
)

# Convert to categorical
y_train_cat = keras.utils.to_categorical(y_train, len(class_names))
y_val_cat = keras.utils.to_categorical(y_val, len(class_names))

print(f"Training samples: {len(X_train)}")
print(f"Validation samples: {len(X_val)}")

# Simple data augmentation
datagen = keras.preprocessing.image.ImageDataGenerator(
    horizontal_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1
)

# Build lightweight model
def build_model(num_classes):
    from tensorflow.keras.applications import MobileNetV2

    base = MobileNetV2(
        include_top=False,
        weights='imagenet',
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        alpha=0.35  # Very small model
    )

    base.trainable = False

    model = keras.Sequential([
        base,
        layers.GlobalAveragePooling2D(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.3),
        layers.Dense(num_classes, activation='softmax', dtype='float32')
    ])

    return model

print("\nBuilding model...")
model = build_model(len(class_names))

model.compile(
    optimizer=keras.optimizers.Adam(LEARNING_RATE),
    loss='categorical_crossentropy',
    metrics=['accuracy'],
    jit_compile=True
)

model.summary()

print("\n" + "="*60)
print("Training (BLAZING FAST with in-memory data!)")
print("="*60 + "\n")

callbacks = [
    keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True, verbose=1),
    keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=2, verbose=1),
    keras.callbacks.ModelCheckpoint('best_model_blazing.keras', save_best_only=True, verbose=1)
]

history = model.fit(
    datagen.flow(X_train, y_train_cat, batch_size=BATCH_SIZE),
    validation_data=(X_val, y_val_cat),
    epochs=EPOCHS,
    callbacks=callbacks,
    verbose=1,
    steps_per_epoch=len(X_train) // BATCH_SIZE
)

# Save model
model.save('food_freshness_blazing.keras')

# Evaluate
print("\n" + "="*60)
print("Evaluation")
print("="*60)
loss, acc = model.evaluate(X_val, y_val_cat, verbose=0)
print(f"\nValidation Accuracy: {acc:.4f} ({acc*100:.2f}%)")

# Generate confusion matrix
print("\nGenerating confusion matrix...")
y_pred = model.predict(X_val, verbose=0)
y_pred_classes = np.argmax(y_pred, axis=1)

from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

cm = confusion_matrix(y_val, y_pred_classes)

plt.figure(figsize=(16, 14))
sns.heatmap(cm, annot=True, fmt='d', cmap='RdYlGn',
            xticklabels=class_names,
            yticklabels=class_names,
            linewidths=0.5)
plt.title('Confusion Matrix', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('True Label', fontsize=11, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=11, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=8)
plt.tight_layout()
plt.savefig('confusion_matrix_blazing.png', dpi=300, bbox_inches='tight')
plt.show()

# Classification report
print("\n" + "="*60)
print("Classification Report")
print("="*60 + "\n")
print(classification_report(y_val, y_pred_classes, target_names=class_names, digits=3))

# Plot training history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

acc_hist = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss_hist = history.history['loss']
val_loss = history.history['val_loss']
epochs_range = range(len(acc_hist))

axes[0].plot(epochs_range, acc_hist, 'b-', label='Training', linewidth=2)
axes[0].plot(epochs_range, val_acc, 'r-', label='Validation', linewidth=2)
axes[0].set_title('Model Accuracy', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

axes[1].plot(epochs_range, loss_hist, 'b-', label='Training', linewidth=2)
axes[1].plot(epochs_range, val_loss, 'r-', label='Validation', linewidth=2)
axes[1].set_title('Model Loss', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('training_blazing.png', dpi=300, bbox_inches='tight')
plt.show()

# Test on sample images
print("\n" + "="*60)
print("Testing on Sample Images")
print("="*60 + "\n")

num_samples = 16
sample_indices = np.random.choice(len(X_val), num_samples, replace=False)

fig, axes = plt.subplots(4, 4, figsize=(16, 16))
axes = axes.flatten()

correct = 0
for idx, sample_idx in enumerate(sample_indices):
    img = X_val[sample_idx]
    true_label = y_val[sample_idx]

    pred = model.predict(np.expand_dims(img, axis=0), verbose=0)
    pred_label = np.argmax(pred)
    confidence = pred[0][pred_label]

    is_correct = (pred_label == true_label)
    correct += int(is_correct)

    axes[idx].imshow(img)
    color = 'green' if is_correct else 'red'
    axes[idx].set_title(f"True: {class_names[true_label]}\nPred: {class_names[pred_label]}\n{confidence:.1%}",
                       fontsize=9, color=color, fontweight='bold')
    axes[idx].axis('off')

plt.suptitle(f'Sample Predictions ({correct}/{num_samples} correct)', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('sample_predictions_blazing.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n" + "="*60)
print("Training Complete!")
print("="*60)
print(f"\nFinal Training Accuracy: {acc_hist[-1]:.4f} ({acc_hist[-1]*100:.2f}%)")
print(f"Final Validation Accuracy: {val_acc[-1]:.4f} ({val_acc[-1]*100:.2f}%)")
print(f"Best Validation Accuracy: {max(val_acc):.4f} ({max(val_acc)*100:.2f}%)")
print(f"Sample Test Accuracy: {correct}/{num_samples} = {correct/num_samples:.1%}")

print("\nFiles saved:")
print("  ✓ food_freshness_blazing.keras (model)")
print("  ✓ best_model_blazing.keras (best checkpoint)")
print("  ✓ training_blazing.png (training curves)")
print("  ✓ confusion_matrix_blazing.png (confusion matrix)")
print("  ✓ sample_predictions_blazing.png (sample predictions)")

print("\n" + "="*60)
print("Note: This model trained on ~10k images for speed.")
print("For best accuracy, use the full dataset (slower training).")
print("="*60)
```
